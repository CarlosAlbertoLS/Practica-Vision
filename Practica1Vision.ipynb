```python
#Se importan las librerias necesarias
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#Lectura de la imagen 1
img = cv2.imread("./kt.jpeg")
print(img.shape)
cv2_imshow(img)

#Lectura de la imagen 2
img2 = cv2.imread("./java.jpg")
print(img2.shape)
cv2_imshow(img2)
```

```python
#Se transforman las imagenes a tipo float
imgf = img.astype(np.float32)
img2f = img2.astype(np.float32)
```

```python
#Suma
suma = (imgf + img2f)/2

#Se transforma a entero sin signo de 8 bits
suma.astype(np.uint8)

#Se muestra el resultado de la suma
cv2_imshow(suma)
```

```python
#Resta
resta = (imgf - img2f)/2

#Se transforman a entero sin signo de 8 bits
resta.astype(np.uint8)

#Se muestra el resultado de la resta
cv2_imshow(resta)
```

```python
#Multiplicacion
result = ((img/255.)*(img2/255))*255
cv2_imshow(result)
```

```python
#Division
result = ((img2f/255.)/(imgf/255))*255
cv2_imshow(result)
```

Usando metodos de openCV

```python
#Suma
sum = cv2.addWeighted(img,0.6,img2,0.4,0.5)
cv2_imshow(sum)
```

```python
#Resta
res = cv2.subtract(img,img2)
cv2_imshow(res)
```

```python
#Multiplicacion
mul = cv2.multiply(img,img2,3)

cv2_imshow(mul)
```
